**Summary** For using ARIMA models **`Non-Seasonal`**
1. plot auto-coorelation to know if stationary or not
    - by observing values ,if there is a decay towards 0 apply first difference,then plot again
    - if become stationary ,deterimne which to use AR or MA or both
    - if get positive correlation at lag-1 then use `AR`
    - if get negative correlation at lag-1 then use `MA`
---
2. if not sure about ACF Autocorrelation function then plot PACF(partial)
    -  if PACF drops off at lag number k ,so use `AR-k` terms Model
    -  if PACF drops off more gradually  ,so use `MA` Model
      ***
    I. detemine how many `AR` terms needed ,this calculated by `ACF` and `PACF` 
    - `AR` Model act like linear regression model and forecast our variable using a combination of past values

    II. if using `MA` if there is jumps ,smooth it and lag for it
    - MA series is usually negatively autocorrelated at lag-1
 
3. `I` Integrated component

   - number of times we have to difference time series to be stationary
   - I=1 means first difference
   > so alltogether:
        - `AR` = 1 
        - `I`  = 1
        - `MA` = 0
        - so we denote our model as ARIMA(1,1,0)

**Summary** For using ARIMA models **`Seasonal`**

used when when time series shows signs of seasonality
- ARIMA(p,d,q)(P,D,Q)m
- where m is the number of periods in each season
- P,D,Q refers to seasonal  part


- seasonal difference is a series of changes from one season to the next
- For Monthly data:
    - `m` will be 12
    - the seasinal difference of any period is the difference between itself and 12 periods prior
- Seasonal models: ca use  non-seasonal differencing, seasonal differencing or both

- if get positive Auto-correlation then use `AR`
- if get negative Auto-correlation then use `MA`

- when to use `AR`
    -  If ACF drops off more gradually
    -  And  PACF drops off at 0 

- when to use `MA`
    -  If ACF drops off at 0 
    -  if PACF drops off more gradually

## In forecasting
- Confidence Interval:
An Interval whithin which we expect our forecast to lie
- we need to specify confidence intervals on which we want our model to present values that lie within the upper and lower bounds

## Model evaluation
1. get residuals -->difference between observation and forecasted value
2. Residuals should be uncorrelated ,so plot in ACF
    - if no Auto correlation so we used all info in model construction
    - If many lags are correlated ,add term to the model to adjust
3. Residuals should have mean close to 0
    - if other than 0 ,model is biased
    - so add the mean to all forcasted values
